```
import pandas as pd
import json
import urllib.request

# PREPARAÇÃO DO DATAFRAME DE REFERÊNCIA 

def load_data():
    url = "https://brasil.io/dataset/covid19/caso?format=csv"
    df  = pd.read_csv(url)
    df  = df.fillna(0)
    df  = df.loc[df["city"] != "Importados/Indefinidos"]
    return df

def cidades():
    url_cidades = "https://raw.githubusercontent.com/leo-smi/latitude-e-longitude-estados-e-cidades-brasileiras/master/cidades.csv"
    df_cidades  = pd.read_csv(url_cidades)
    return df_cidades
```

```
# Carregamento dos dados
df_data    = load_data()
df_cidades = cidades()

# CIDADES COM CASOS
# Dicionário vazio
cidades_com_casos = {"cidade":[]}

for index, row in df_data.iterrows():
    cidade = row["city"]
    if  cidade != 0: 
        if cidade not in cidades_com_casos["cidade"]:
            cidades_com_casos["cidade"].append(cidade)

df_comCasos = pd.DataFrame(cidades_com_casos)

# CIDADES SEM CASOS
cidades_sem_casos = {"cidade":[], "latitude":[], "longitude":[]}

for index, row in df_cidades.iterrows():
    cidade = row["nome"]
    lat    = row["latitude"]
    lng    = row["longitude"]
    if cidade not in cidades_com_casos["cidade"]:
        cidades_sem_casos["cidade"].append(cidade)
        cidades_sem_casos["latitude"].append(lat)
        cidades_sem_casos["longitude"].append(lng)

df_semCasos = pd.DataFrame(cidades_sem_casos)
df_semCasos.to_csv("cidades_sem_casos_de_covid19_" + str(max(df_data["date"])) + ".csv")
df_semCasos
```
